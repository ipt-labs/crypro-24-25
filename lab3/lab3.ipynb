1. Реалізувати підпрограми із необхідними математичними операціями: обчисленням оберненого елементу за модулем із використанням розширеного алгоритму
Евкліда, розв’язуванням лінійних порівнянь. При розв’язуванні порівнянь потрібно коректно обробляти випадок із декількома розв’язками, повертаючи їх усі.

Очислення оберненого елементу за модулем із використанням розширеного алгоритму Евкліда

```python
# a^(-1) mod m
def euclidean_algorithm(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        gcd, x, y = euclidean_algorithm(b % a, a)
        return (gcd, y - (b // a) * x, x)

def mod_inverse(a, m):
    gcd, x, y = euclidean_algorithm(a, m)
    if gcd != 1:
        raise ValueError('Обернене за модулем не існує')
    else:
        return x % m
a = 2
m = 5
try:
    inverse = mod_inverse(a, m)
    print("Обернене число для", a, "за модулем", m, "дорівнює", inverse)
except ValueError as e:
    print(e)
```

Розв’язування лінійних порівнянь

```python
# a x ≡ b (mod m)
def congruence(a, b, m):
    gcd, x0, _ = euclidean_algorithm(a, m)
    solutions = []
    if b % gcd == 0:
        x0 = (x0 * (b // gcd)) % m
        for i in range(gcd):
            solutions.append((x0 + i * (m // gcd)) % m)
        return solutions
    else:
        raise ValueError("Рівняння не має розв'язку")

a = 110
b = 3
m = 5

try:
    solutions = congruence(a, b, m)
    print(f"Розв'язки порівняння {a}x ≡ {b} (mod {m}):", solutions)
except ValueError as e:
    print(e)
```

2. За допомогою програми обчислення частот біграм, яка написана в ході виконання комп’ютерного практикуму №1, знайти 5 найчастіших біграм запропонованого
шифртексту (за варіантом).

```python
from collections import Counter
import re  

def read_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().strip()
    text = re.sub(r'[^а-яА-Я]', '', text)  
    return text

def calculate_bigram_frequency(text):
    bigrams_without_overlap = [text[i:i+2] for i in range(0, len(text) - 1, 2)]
    total_bigrams_without_overlap = len(bigrams_without_overlap)  
    freq_without_overlap = Counter(bigrams_without_overlap)
    freq_without_overlap = {bigram: count / total_bigrams_without_overlap for bigram, count in freq_without_overlap.items()}

    return freq_without_overlap

text = read_text('./04.txt')

bigram_freq_without_overlap = calculate_bigram_frequency(text)
bigrams = dict(sorted(bigram_freq_without_overlap.items(), key=lambda item: item[1], reverse=True)[:5])

print("Найчастіші біграми ШТ:")
for bigram, freq in bigrams.items():
    print(f"{bigram}: {freq}")
```

3. Перебрати можливі варіанти співставлення частих біграм мови та частих біграм шифртексту (розглядаючи пари біграм із п’яти найчастіших). Для кожного співставлення знайти можливі кандидати на ключ (a,b) шляхом розв’язання системи.

```python
alphabet = "абвгдежзийклмнопрстуфхцчшщьыэюя"
m = len(alphabet) 
m2 = m ** 2  

# Cловники для перетворення символів у числа та назад
alphabet_to_index = {char: i for i, char in enumerate(alphabet)}
index_to_alphabet = {i: char for i, char in enumerate(alphabet)}

# Перетворення біграми у числове значення
def bigram_to_number(bigram):
    return alphabet_to_index[bigram[0]] * m + alphabet_to_index[bigram[1]]

# Знаходження можливих ключів (a, b)
def k(ct, vt):
    keys = set()
    cipher_num = [bigram_to_number(bigram) for bigram in ct]
    plain_num = [bigram_to_number(bigram) for bigram in vt]

    for i, y1 in enumerate(cipher_num):
        for j, y2 in enumerate(cipher_num):
            if i == j:
                continue

            delta_Y = (y1 - y2) % m2
            for k, x1 in enumerate(plain_num):
                for l, x2 in enumerate(plain_num):
                    if k == l:
                        continue

                    delta_X = (x1 - x2) % m2
                    try:
                        a_candidates = congruence(delta_X, delta_Y, m2)
                    except ValueError:
                        continue

                    for a in a_candidates:
                        b = (y1 - a * x1) % m2
                        keys.add((a, b))
    return list(keys)

ct = ['еш', 'еы', 'шя', 'ск', 'до']
vt = ['ст', 'но', 'то', 'на', 'ен']
keys = k(ct, vt)
print("Кандидати на ключі (a, b):")
for a, b in keys:
    print(f"(a={a}, b={b})")
```

4. Для кожного кандидата на ключ дешифрувати шифртекст. 


```python
# Перетворення числового значення у біграму
def number_to_bigram(number):
    first_char = index_to_alphabet[number // m]
    second_char = index_to_alphabet[number % m]
    return first_char + second_char

# Функція дешифрування тексту
def decrypt_text(ciphertext, key):
    a, b = key
    a_inv = mod_inverse(a, m2)  
    plaintext = []

    for i in range(0, len(ciphertext), 2):
        bigram = ciphertext[i:i+2]
        y = bigram_to_number(bigram)
        x = (a_inv * (y - b)) % m2
        plaintext.append(number_to_bigram(x))

    return ''.join(plaintext)

text = read_text('./04.txt')
print("Результати дешифрування:")
for key in keys:
    try:
        decrypted_text = decrypt_text(text, key)
        print(f"(a={key[0]}, b={key[1]}): {decrypted_text}")
    except ValueError as e:
        print(f"(a={key[0]}, b={key[1]}): неможливо дешифрувати - {e}")
```

Якщо шифртекст не є змістовним текстом російською мовою, відкинути цього кандидата.

```python
# Список заборонених біграм
forbidden_b = ['аь', 'юь', 'еь', 'оь']

# Функція для перевірки на заборонені біграми
def zaboroneni_bigrams(text):
    for bigram in forbidden_b:
        if bigram in text:
            return True 
    return False  

print("Результати дешифрування:")
for key in keys:
    decrypted_text = decrypt_text(text, key)  
    if not zaboroneni_bigrams(decrypted_text):
        print(f"(a={key[0]}, b={key[1]}): {decrypted_text}")  
    else:
        print(f"(a={key[0]}, b={key[1]}): містить заборонені біграми")
```
