## розширений алгоритм Евкліда

```
def extended_euclid(a, b):
    if b == 0:
        return a, 1, 0
    
    gcd, x1, y1 = extended_euclid(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return gcd, x, y

def modulo_inverse(a,m):
    if extended_euclid(a,m)[0]==1:
        return extended_euclid(a,m)[1]
    else:
        return 0



print(extended_euclid(20, 52))
print(modulo_inverse(20,52))
print(modulo_inverse(2,5))
```

## Розв'язування лінійних порівнянь

```
def calc_linear_equation(a, b, n):
    d=extended_euclid(a,n)[0]
    if d == 1:
        return (extended_euclid(a,n)[1] * b ) % n
    elif b%d ==0:
        a1=a//d
        b1=b//d
        n1=n//d
        x0=calc_linear_equation(a1,b1,n1)
        return [x0+i*n1 for i in range(0,d)]
    else:
        return 0
print(f'Приклад з однією відповіддю (20x=3 mod52): {calc_linear_equation(20, 3, 52)}')
print(f'Приклад з однією відповіддю (17x=3 mod23): {calc_linear_equation(17, 3, 23)}')
print(f'Приклад з декількома відповідями (20x=16 mod52): {calc_linear_equation(20, 16, 52)}')
```

## Частота біграм

Код з лаб 1

```
def find_bigram_frequency_and_count(text:str, alphabet:str, cross:bool):
    bigram_count_dict={}
    bigram_frequency_dict={}
    if cross:
        for i in range(len(text)-1):
            key=text[i]+text[i+1]
            bigram_count_dict[key] = bigram_count_dict.get(key,0) + 1
        for key in bigram_count_dict.keys():
            bigram_frequency_dict[key]=round(bigram_count_dict[key]/(len(text)-1),5)
    if not cross:
        for i in range(0,len(text)-1,2):
            key=text[i]+text[i+1]
            bigram_count_dict[key] =bigram_count_dict.get(key,0) + 1
        for key in bigram_count_dict.keys():
            bigram_frequency_dict[key]=round(bigram_count_dict[key]/(len(text)/2),5) 
    return  bigram_count_dict, bigram_frequency_dict

def find_letter_frequency_and_count(text:str,alphabet:str):

    letter_count_dict={}
    for c in text:
        letter_count_dict[c]=letter_count_dict.get(c,0)+1

    letter_frequency_dict={}
    for c in letter_count_dict:
        letter_frequency_dict[c]=round(letter_count_dict[c]/len(text),5)
    return letter_count_dict, letter_frequency_dict

```

Читаємо текст з файлу, редагуємо, створюємо біграми та обираємо топ 5

```
text=open('03.txt','r').read().replace('n','').replace('\n','')
alphabet = 'абвгдежзийклмнопрстуфхцчшщьыэюя'
b_c,b_f = find_bigram_frequency_and_count(text, alphabet, False)
print(len(text))
print(sum(b_c.values()))
top_5_bigrams = sorted(b_c, key=b_c.get, reverse=True)[:5]

for i in top_5_bigrams:
    print(i, b_c[i])
```

Пари популярних біграм для майбутнього пошуку a і b 

```
def pairs_of_the_most_common(lst,state_pair=True):
    pairs = []
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)): 
            pairs.append((lst[i], lst[j]))
    return pairs
text_popular_pairs=pairs_of_the_most_common(top_5_bigrams)
print(text_popular_pairs)
lang_popular_pairs=pairs_of_the_most_common(['ст','но', 'то', 'на', 'ен'])
print(lang_popular_pairs)
```

Перетворюємо біграми на числа (як в методичці)

```
def pairs_nums(list,state_pair=True):
    global alphabet
    pairs=[]
    if state_pair:
        for pair in list:
            x1=alphabet.index(pair[0][0])*31+alphabet.index(pair[0][1])
            x2=alphabet.index(pair[1][0])*31+alphabet.index(pair[1][1])
            pairs.append((x1,x2))
    else:
        for i in list:
            x1=alphabet.index(i[0])*31+alphabet.index(i[1])
            pairs.append(x1)
    return pairs
text_num_pairs=pairs_nums(text_popular_pairs)
lang_num_pairs=pairs_nums(lang_popular_pairs)
print(text_num_pairs)
print(lang_num_pairs)
```

Шукаємо а і b (в методичці пункт про атаку на аффіний шифр)

```
def potential_keyss(pair1,pair2):
    x1=pair1[0]-pair1[1]
    y1=pair2[0]-pair2[1]
    a=calc_linear_equation(x1,y1,31**2)
    if a == 0:
        return 'no a^-1'
    elif isinstance(a,int):
        b=(pair2[0]-a*pair1[0])%31**2
        return a,b
    else:
        b=(pair2[0]-a[0]*pair1[0])%31**2
        return a[0],b
pot_keys=[]
for i in lang_num_pairs:
    for j in text_num_pairs:
        pot_keys.append(potential_keyss(i,j))
print(pot_keys)
```

дешифруємо тексти

```
def make_bigrams_list(str, alphabet):
    my_list = []
    i = 0
    while i < len(str) - 1:
        key = str[i] + str[i + 1]
        my_list.append(key)
        i = i + 2
    return my_list
def dcd(text,key):
    if key == 'no a^-1':
        return None
    
    decrypt_str = ''
    for j in range(0, len(text)):
        x = int(modulo_inverse(key[0], 31 ** 2)) * (text[j] - key[1]) % (31 ** 2)
        x1 = x // 31
        x2 = x % 31
        decrypt_str += alphabet[x1] + alphabet[x2]
        j = j + 1
    return decrypt_str
text_in_bigrams=make_bigrams_list(text,alphabet)
bigram_nums=pairs_nums(text_in_bigrams,False)
print(bigram_nums)
decoded_texts={}

for key in pot_keys:
    decoded_texts[key]=dcd(bigram_nums,key)
print(decoded_texts)
```

Порівнюємо декодовані шифри з характеристиками рос мови (найпопулярніші букви/біграми)

```
keys_pop_letters={}
for key in list(decoded_texts.keys()):
    if decoded_texts[key]:
        d_c,_ = find_letter_frequency_and_count(decoded_texts[key], alphabet)
        t5 = sorted(d_c, key=d_c.get, reverse=True)[:5]
        l5 = sorted(d_c, key=d_c.get, reverse=True)[-5:]
        if len(np.intersect1d(t5,['о','а','е']))>1 and len(np.intersect1d(l5,['ф','щ','ь']))>1:
            keys_pop_letters[key]=[t5,l5]
print(f'{keys_pop_letters=}')

keys_pop_bigrams={}
for k in list(decoded_texts.keys()):
    if decoded_texts[k]:
        kk_c,_=find_bigram_frequency_and_count(decoded_texts[k],alphabet,True)
        t5_b = sorted(kk_c, key=kk_c.get, reverse=True)[:6]
        if len(np.intersect1d(t5_b,['ст', 'но', 'ен', 'то', 'на']))>2:
            keys_pop_bigrams[k]=t5_b
print(f'{keys_pop_bigrams=}')
correct_key=np.intersect1d(list(keys_pop_bigrams.keys()),list(keys_pop_letters.keys()))
print(f'{correct_key=}')
```
