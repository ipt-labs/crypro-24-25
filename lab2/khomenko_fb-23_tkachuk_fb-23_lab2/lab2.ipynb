# Розшифрування тексту за варінтом

функція з минулої лр

```
def letter_frequency(text:str):
    letter_count_dict={}
    for c in text:
        letter_count_dict[c]=letter_count_dict.get(c,0)+1
    letter_frequency_dict={}
    
    for c in letter_count_dict:
        letter_frequency_dict[c]=round(letter_count_dict[c]/len(text),5)
    return letter_count_dict, letter_frequency_dict
```

## Індекс відповідності 

для визначення істинного r (довжини ключа) використаємо перший алгоритм з методичних вказівок

1) Для кожного кандидата r=2,3.... розбити шифротекст Y на блоки Y_1,Y_2...Y_r.

2)Обчислити значення індексу відповідності для кожного блоку

3) Якщо сукупність одержаних значень (буде використовуватись середнє значення індексів відповідності блоків при певній довжині ключа) схиляється до теоретичного значення I для даної мови, то значення r вгадане правильно. Значення I для рос. мови = 0,055

функція block_creation розділяє текст на блоки. береться символ через keylength символів.

Функція index_of_coincidence рахує значення індексу відповідності згідно формули в методичних вказівках

Функція find_block_I перебирає ключі розміром 2-32, розбиває на блоки, рахує індекс відповідності

```
def blocks_creation(text,keylength):
    blocks=[]
    i=0
    while i < keylength:
        block=""
        j=0
        while i+j*keylength < len(text):
            block+=text[i+j*keylength]
            j+=1
        blocks.append(block)
        i+=1
    return blocks

def index_of_coincidence(text):
    text=sorted(text)
    l_c, _ = letter_frequency(text)
    sum = 0
    for i in l_c.keys():
        sum+=l_c[i]*(l_c[i]-1)
    index=sum/(len(text)*(len(text)-1))
    return index

def find_block_I(text):
    for key in range(2,32):
        i=0
        index=[]
        blocks=blocks_creation(text,key)
        while i < len(blocks):
            index.append(index_of_coincidence(blocks[i]))
            i+=1
        new_row={'Key Length': key, "Index":np.mean(index)}
        keys_indexes.loc[len(keys_indexes)]=new_row

keys_indexes=pd.DataFrame(columns=['Key Length',"Index"])
find_block_I(text)
keys_indexes.to_excel('keys_indexes.xlsx')
keys_indexes
```

У ключа з довжиною 14 середній індекс відповідності = 0,05667, що схиляється до теоретичного I рос. мови, тому робимо висновок, що довжина ключа =14.

Тепер знаходимо ключ. З минулої лабораторної відомо, що "найпопулярніші" букви рос алфавіту це оеиа

```
def find_key(text):
    blocks=blocks_creation(text,14)
    for block in blocks:
        _,f=letter_frequency(block)
        letter_maxfreq=max(f,key=f.get)
        possible_letter1=alphabet[(alphabet.index(letter_maxfreq) - alphabet.index('о'))%32]
        possible_letter2=alphabet[(alphabet.index(letter_maxfreq) - alphabet.index('е'))%32]
        possible_letter3=alphabet[(alphabet.index(letter_maxfreq) - alphabet.index('и'))%32]
        possible_letter4=alphabet[(alphabet.index(letter_maxfreq) - alphabet.index('а'))%32]
        print(possible_letter1,possible_letter2,possible_letter3,possible_letter4)
find_key(text)
```

оскільки ключ повинен бути змістовним, то це экомаятникфуко

```
def decode(text,key):
    decoded_text=""
    i=0
    for c in text:
        decoded_text+=alphabet[(alphabet.index(c) - alphabet.index(key[i%len(key)]))%32]
        i+=1
    return decoded_text
decoded_text=decode(text,'экомаятникфуко')
print(decoded_text)
with open('decoded_text.txt', 'w') as file:
    file.write(decoded_text)
```

# Шифрування тексту

Був взятий вірш "У лукоморья дуб зелёный" Пушкіна

```
import matplotlib.pyplot as plt
def encode(text,key):
    encoded_text=""
    i=0
    for c in text:
        encoded_text+=alphabet[(alphabet.index(c) + alphabet.index(key[i%len(key)]))%32]
        i+=1
    with open(f'encoded_text_{len(key)}.txt', 'w') as file:
        file.write(encoded_text)
    return encoded_text


raw_text=open('text.txt','r').read().replace('n','').replace('\n','').replace('ё','е').lower()
text_to_encode=''.join(char for char in raw_text if char in alphabet)

text_to_encode_df=pd.DataFrame(columns=['Розмір ключа','Індекс відповідності'])
new_row={'Розмір ключа': 0, "Індекс відповідності": index_of_coincidence(text_to_encode)}
text_to_encode_df.loc[len(text_to_encode_df)]=new_row

keys=["ад","кпи","цирк","пекло","страждання", "вдохновение", "драконийклад","волшебныемиры","царицавцарстве",
"вортурастетгриб", "любилизапахдождя", "стихсостоитизрифм", "будуждатьтебяутром", "учитьсянужнозаранее", "помогитемнепридумать"]
for key in keys:
    cyphertext=encode(text_to_encode,key)
    new_row={'Розмір ключа': len(key), "Індекс відповідності": index_of_coincidence(cyphertext)}
    text_to_encode_df.loc[len(text_to_encode_df)]=new_row
text_to_encode_df.to_excel('text_to_encode.xlsx')
text_to_encode_df
```
